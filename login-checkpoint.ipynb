```python
import mysql.connector
from tkinter import *
import os

# Function to register a new user
def register():
    username = username_entry.get()
    password = password_entry.get()

    # Connect to MySQL database
    conn = mysql.connector.connect(
        host="localhost",
        user="root",
        password="",
        database="py_lg_rg_db"
    )
    cursor = conn.cursor()

    # Check if username already exists
    cursor.execute("SELECT * FROM users WHERE username = %s", (username,))
    result = cursor.fetchone()
    if result:
        status_label.config(text="Username already exists")
    else:
        # Insert new user into the database
        cursor.execute("INSERT INTO users (username, password) VALUES (%s, %s)", (username, password))
        conn.commit()
        status_label.config(text="Registration successful")

    cursor.close()
    conn.close()

# Function to perform login
def login():
    username = username_entry.get()
    password = password_entry.get()

    # Connect to MySQL database
    conn = mysql.connector.connect(
host="localhost",
        user="root",
        password="",
        database="py_lg_rg_db"
    )
    cursor = conn.cursor()

    # Check if username and password match
    cursor.execute("SELECT * FROM users WHERE username = %s AND password = %s", (username, password))
    result = cursor.fetchone()
    if result:
        status_label.config(text="Login successful")
        # Open the finalDetection.ipynb file
        os.system('jupyter notebook finalDetection.ipynb &')
    else:
        status_label.config(text="Invalid username or password")

    cursor.close()
    conn.close()

# GUI setup
root = Tk()
root.title("Login System")

username_label = Label(root, text="Username:")
username_label.grid(row=0, column=0, padx=10, pady=5)
username_entry = Entry(root)
username_entry.grid(row=0, column=1, padx=10, pady=5)

password_label = Label(root, text="Password:")
password_label.grid(row=1, column=0, padx=10, pady=5)
password_entry = Entry(root, show="*")
password_entry.grid(row=1, column=1, padx=10, pady=5)

register_button = Button(root, text="Register", command=register)
register_button.grid(row=2, column=0, padx=10, pady=5)

login_button = Button(root, text="Login", command=login)
login_button.grid(row=2, column=1, padx=10, pady=5)

status_label = Label(root, text="")
status_label.grid(row=3, columnspan=2, padx=10, pady=5)

root.mainloop()
```
